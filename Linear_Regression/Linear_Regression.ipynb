```python
import numpy as np
import random
import matplotlib.pyplot as plot
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
plot.scatter(x,y)
```

```python
# y_pred = wx+b   w is slope and b is the intercept
w = np.random.rand(1)
b= np.random.rand(1)
print(w)
print(b)
def y_pred(w,b,x): # Linear Regression Function
    return w*(x)+b
```

```python
plot.scatter(x,y)
plot.plot(x,y_pred(w,b,x))
```

```python
Loss = np.mean(np.square(y-y_pred(w,b,x))) # Loss = (y-ypred)^2
print(np.square(y-y_pred(w,b,x)))
print(Loss)
```

```python
# Gradient Descent
# w = w -alpha*dL/dw  = w-alpha*2*(wx+b-y)*x
# b = b -alpha*dL/db
# L = (y_pred-y)^2
# L = (wx+b - y)^2
# dL/dw  = 2*(wx+b-y)*x
# dL/db = 2*(wx+b-y)

alpha = 0.01
# for(int i=0;i<1000;i++) java and c would look like this
for i in range(10):

  w_grad = np.mean(2*(y_pred(w,b,x)-y)*x)
  w = w - alpha*w_grad
  b_grad = np.mean(2*(y_pred(w,b,x)-y))
  b  = b -alpha*w_grad
  Loss = np.mean(np.square(y-y_pred(w,b,x))) # Loss = (y-ypred)^2
  print(w_grad)
plot.scatter(x,y)
plot.plot(x,y_pred(w,b,x))
plot.show()



```
